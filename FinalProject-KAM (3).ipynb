# Best Pittsburgh Neighborhood Based on City Designated Historical Sites

This notebook attempts to determine which Pittsburgh neighborhood is the best. The analysis is based on the number of city-designated individual historical sites. 

## Analysis

This code is designed to download the data from the city-designated historical sites downloaded from Western Pennsylvania Regional Data Center.

```python
import pandas as pd
%matplotlib inline
```

```python
data = pd.read_csv("historicalSites3.csv")
data1 = data.drop(columns=['Unnamed: 1', 'Unnamed: 2', 'Unnamed: 3', 'Unnamed: 4', 'Unnamed: 5', 'Unnamed: 6', 'Unnamed: 7', 'Unnamed: 9', 'Unnamed: 10', 'Unnamed: 11', 'Unnamed: 12', 'Unnamed: 13'])
data2 = data1.rename(columns = {'Unnamed: 8' : 'Neighborhood'})
data2.head(5)
```

To determine which neighborhood is the best, I tallied up each the counts for each neighborhood.

```python
nCount = data2.groupby('Neighborhood').count()
nCount
```

From this data, I can conclude what the top neighborhoods in Pittsburgh are. The top three neighborhoods with the highest number of city-designated historical sites are the Central Business District, Lower Lawrenceville, and Shadyside.

Lower Lawrenceville and Shadyside are both tied with 8 city-designated historical sites. The Central Business District, also known as Downtown, has the most city-designated historical sites, with 11 in total.

Therefore, according to my analysis, the Central Business District is the best Pittsburgh neighborhood to live in. My metric is based on tourism that is likely to occur in that area. If an area has more tourism, it means that more money is being injected into the local economy.
