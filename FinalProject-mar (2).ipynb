# Best Pittsburgh Neighborhood Based on the Air Quality

```python
import pandas as pd

air_quality = pd.read_csv("/ihome/cmpinf0010_2025f/mar879/Final Project-mar879/allegheny_air_quality.csv")

air_quality.head()
```

```python
pm25 = air_quality[air_quality["parameter"] == 'PM25']
```

```python
pm25_avg = pm25.groupby('site')['report_value'].mean().reset_index()
```

```python
pm25_avg = pm25_avg.sort_values('report_value')
```

```python
pm25_avg.head(10)
```

```python
pm25 = air_quality[air_quality["parameter"].str.contains("PM2.5|PM25", case=False, na=False)]

pm25_summary = pm25.groupby('site')['report_value'].agg(Avg_PM25 = 'mean', Min_PM25 = 'min', Max_PM25 = 'max', Count = 'count').reset_index()

pm25_summary = pm25_summary.sort_values('Avg_PM25')

pm25_summary['Rank'] = range(1, len(pm25_summary)+1)

pm25_summary = pm25_summary[['Rank', 'site', 'Avg_PM25', 'Min_PM25', 'Max_PM25', 'Count']]

pm25_summary = pm25_summary[pm25_summary['site'] != 'Lawrenceville 2']

pm25_summary = pm25_summary.sort_values('Avg_PM25')  

pm25_summary['Rank'] = range(1, len(pm25_summary) + 1)

pm25_summary
```
