# Best Pittsburgh Neighborhood Based on City Designated Historical Sites

This notebook attempts to determine which Pittsburgh neighborhood is the best. The analysis is based on the number of city-designated individual historical sites. Historical sites are known to generate tourism and activity in their surrounding areas. These activities stimulate economic activity, thereby increasing the amount of money in circulation in that area. A neighborhood with more city-designated historical sites has more economic activity, therefore making it a better place to live. With that in mind, this analysis is done to find the neighborhood with the most city-designated historical sites, as it directly correlates to tourism and economic activity.

### Analysis

This code is designed to download the data from the city-designated historical sites downloaded from Western Pennsylvania Regional Data Center. It is then organized, where all of the unnecessary columns are dropped. This just leaves the "Neighborhood" column, which lists each neighborhood for each time that it has a city-designated historical site. For example, if Shadyside has four historical sites, its name would appear four times in the list.

```python
import pandas as pd
```

```python
data = pd.read_csv("cityHistoricalSites(1).csv") ## Import Data
data1 = data.drop(columns=['objectid', 'globalid', 'historic_i', 'lotblock', 'street', 'address', 'shape_leng_1', 'alternativ', 'name', 'provided_a', 'shape_area', 'shape_length'])
data1.head(5) ## Drop the columns that are not necessary and print out the first 5 observations
```

After importing the data, I then made a dictionary, for which the code is shown below. By making it into a dictionary, each neighborhood is listed, and then it has a key with it, which shows the number of times it appears in the column. Then, it is organized based on the highest number of appearances. The neighborhood that appears the most in the dictionary is listed first, followed by the second most, and so forth. The dictionary is then printed, and its output is shown below.

```python
## Make a dictionary
dictionary = dict()

for index, row in data1.iterrows(): ## Run through each row
    item = row["Neighborhood"] ## Get the item
    if item not in dictionary: ## If not in dictionary...
        dictionary[item] = 1 ## Add it
    else: ## If in dictionary...
        dictionary[item] += 1 ## Add one to the tally

## Sort the dictionary in numerical order
dictionary = sorted(dictionary.items(), key = lambda item: item[1], reverse = True)

## Print out the dictionary
print(dictionary)
```

### Visuals

Once the dictionary was established, it was then necessary to make visual representations of the data. Two bar graphs were used for this step. First, a data frame was made of the first seventeen observations in the dictionary. Because it is sorted numerically, these observations represent the Pittsburgh neighborhoods with the most city-designated historical sites. The data frame is printed below. Underneath this code, the bar graph is shown.

```python
import matplotlib.pyplot as plt

## Put the sorted data into a data frame and take the top 17 observations
df = pd.DataFrame(dictionary, columns = ["Neighborhood", "Count"]).head(17)
print(df)
```

```python
## Make a bar graph that shows the top 17 neighborhoods
graph = df.plot.bar(x="Neighborhood", y = "Count", title = "Top 17 Pittsburgh Neighborhoods", colormap='Paired')
graph.set_ylabel("Number of City-Designated Historical Sites")
plt.show()
```

The code below is relatively the same as above; however, it shows the bottom thirty-five neighborhoods based on city-designated historical sites. Again, the data frame is printed out and then the bar graph is shown below.

```python
## Put the sorted data into a data frame and take the last 35 observations
df = pd.DataFrame(dictionary, columns = ["Neighborhood", "Count"]).tail(35)
print(df)
```

```python
## Make a bar graph that shows the top 10 neighborhoods
bottomGraph = df.plot.bar(x="Neighborhood", y = "Count", title = "Bottom 35 Pittsburgh Neighborhoods", colormap='Paired')
bottomGraph.set_ylabel("Number of City-Designated Historical Sites")
plt.show()
```
