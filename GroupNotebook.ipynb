### Conclusion

Our final metric to determine the best Pittsburgh neighborhoods is overall quality of life. After analyzing data from all of the Pittsburgh neighborhoods and looking at multiple different metrics, we determined that East Liberty is the best neighborhood to live in. This was based on multiple different factors. First, we analyzed the average air quality in each neighborhood. In the dataset, East Liberty had the best overall air quality for all of the neighborhoods. Secondly, we analyzed waste recovery locations. We rationalized that this would help us determine which neighborhoods are the cleanest by having the most trash cans and sanitation. Again, East Liberty was at the top of this list. We then analyzed the 2025 dog licenses data set to determine which neighborhoods applied for and received the most dog licenses. The best neighborhoods in this category were found to be Upper Saint Clair, Baldwin, and Avalon, however, East Liberty was in the top fifty percent of the issued dog licenses. Finally, we looked at the data set containing individual city designated individual historical sites. We found that the Central Business District, Lawrenceville, and Shadyside had the highest individual historical sites. East Liberty had 2 historical sites, which are enough to generate tourism to the area. Therefore, we concluded that East Liberty is the best Pittsburgh neighborhood because someone who lives there would have the best quality of life.
